### Introduction 


Assume that, a professor is teaching an online class, in which he has got 7 students; Student1,... Student7. Of these 7 students, 3 of them (Student1, Student2 and Student3) are undegrad.(UG) students while 4 of them (Student4, Student5, Student6 and Student7) are Grad (G) students. 

He wants to create a matrix that represents the students grades. Eventually, he wants the analyze the data in this matrix.

There are 2 exams, 5 Hw assignments and 1 final project in this class. 
Both exams (Exam1, Exam2) are graded out of 100, but the weight of the average of these exams are 40 %. Hws (Hw1,Hw2,....,) are worth graded out of 100 and the weight of the average of these hws are 50 % , and finally the project is graded out of 100, but the weight of the project is 10 %.

The students's final grade is calculated in the following way : (Average of exams * 0.4) +  (Average of Hws * 0.5) + (Project * 0.1).

The student gets "A" if the overall grade is above 90%, "B" if the overall grade is between 80 - 90%, "C" if the overall grade is between 70 - 80%, "D" if the overall grade is between 60 - 70% and finally "D" if the overall grade is below 60%

For example : Let us assume that Student1 has got 80 from Exam1, 90 from Exam2, 100 from Hw1, 80 from Hw2, 90 from Hw3,100 from Hw4, 75 from Hw5 and finally 90 from the Project.

The final grade of Student1 is : ((80+90)/2 * 0.4) + ((100+80+90+100+75)/5 * 0.5) + (90 * 0.1) = 87.5, and ends up with "B".


Our job in the first two questions is to help this instructor to design his matrix, and calculate the final grades and the letter grades if these students.



### Question 1)

- 1a) Create a dataset that contains Test 1 and Test 2 Scores of these students. Please use random integer number creation between 50 and 100 for each test. (Use np.random.seed(2020)) for your randint number generation so that our results match)
- 1b)Perform the same action for Hws.
- 1c)Perform the same action for Project.

(The sample outputs are given below)

```python
## (1a) Exams 
## make sure to use "2022" for your random number generator i.e.:  np.random.seed(2022) 
## so that you can get the same number as I did..


import numpy as np
import pandas as pd
from numpy.random import randint
np.random.seed(2022)

Exams = pd.DataFrame(randint(50,100,size=(7,2)),index='Student1 Student2 Student3 Student4 Student5 Student6 Student7'.split(),columns='Test1 Test2'.split())

Exams
```

```python
## (1b)Hws

## make sure to use "2022" for your random number generator i.e.:  np.random.seed(2022) 
## so that you can get the same number as I did..


np.random.seed(2022)
Hws = pd.DataFrame(randint(50,100,size=(7,5)),index='Student1 Student2 Student3 Student4 Student5 Student6 Student7'.split(),columns='Hw1 Hw2 Hw3 Hw4 Hw5'.split())

Hws
```

```python
## (1c)Project 
## make sure to use "2022" for your random number generator i.e.:  np.random.seed(2022) 
## so that you can get the same number as I did..

np.random.seed(2022)
Project = pd.DataFrame(randint(50,100,size=(7,1)),index='Student1 Student2 Student3 Student4 Student5 Student6 Student7'.split(),columns='Project'.split())
Project
```

```python
## grades_1 (after merging/concatenating)

grades_1 = pd.concat([Exams,Hws,Project], axis=1)
grades_1
```

### Question 2)

- 2a) Calculate the average for the test scores, and create a new variable for the averages. 
- 2b) Perform the same action for Hws.
- 2c) Re-arrange the column order as given below

(The sample outputs are given below)

```python
## (2a) Add Test Averages

grades_1['Test Average'] = (grades_1['Test1'] + grades_1['Test2'])/2
grades_1
```

```python
##(2b) Add hw averages

grades_1['Hw Average'] = (grades_1['Hw1'] + grades_1['Hw2']+ grades_1['Hw3'] + grades_1['Hw4']+grades_1['Hw5'])/5
grades_1
```

```python
## (2c) grades_1(columns are reordered)

grades_2 = grades_1[['Test1', 'Test2','Test Average', 'Hw1', 'Hw2','Hw3', 'Hw4', 'Hw5','Hw Average', 'Project']]   
grades_2
```

### Question 3)

- 3a) Create a variable called "Final Average", with the formula provided at the "Introduction"
- 3b) Create another variable called "Letter Grade", with the condition/formula provided at the "Introduction"
- 3c) The instructor believes that "Test Average" is correlated with "Final Average". Determine the "Test Average" of those students whose "Final Average" is above 80.

(The sample outputs are given below)

```python
## (3a) 

grades_2['Final Average'] = grades_2['Test Average']*0.4 + grades_2['Hw Average']*0.5 + grades_2['Project']*0.1
grades_2
```

```python
## (3b)

def f(row):
    if row['Final Average'] > 90:
        val = 'A'
    elif row['Final Average'] > 80:
        val = 'B'
    elif row['Final Average'] > 70:
        val = 'C'
    elif row['Final Average'] > 60:
        val = 'D'
    else:
        val = "F"
    return val
```

```python
grades_2['Letter Grade'] = grades_2.apply(f, axis=1)
grades_2
```

```python
## (3c)

grades_2[grades_2['Final Average']> 80]['Test Average']
```
